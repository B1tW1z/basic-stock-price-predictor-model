A super basic ML model to predict Stock prices 

Importing yfinance
 
**yfinance**: A Python library that allows users to access financial data from Yahoo Finance, enabling easy retrieval of stock prices, historical data, and other market information for analysis and trading.

In yfinance, a ticker refers to the unique symbol that identifies a publicly traded company or asset on the stock market (e.g., "GOOG" for Alphabet Inc Class C). It’s used to fetch data related to that specific company, such as stock prices, financials, and historical data.

```python
import yfinance as yf
```

```python
goog = yf.Ticker("GOOG")
```

Fetching Stock information

```python
goog.info
```

```python
# All time data for stock
goog = goog.history(period="max")
```

```python
goog
```

```python
goog.index
```

```python
goog.plot.line(y="Close", use_index=True)
```

```python
goog["Tomorrow"] = goog["Close"].shift(-1)
```

```python
goog
```

```python
goog["Target"] = (goog["Tomorrow"] > goog["Close"]).astype(int)
```

```python
goog
```

A random classifier is a simple machine learning model that makes predictions based on random guessing. It assigns class labels to instances without considering any input features, often serving as a baseline for evaluating the performance of more complex classifiers. Its predictions are typically based on the distribution of classes in the training data.

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = goog.iloc[:-100]
test= goog.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

predictions = model.predict(test[predictors])
```

```python
import pandas as pd 

predictions = pd.Series(predictions, index=test.index)
```

```python
precision_score(test["Target"], predictions)
```

```python
combined = pd.concat([test["Target"], predictions], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model): 
        model.fit(train[predictors], train["Target"])
        predictions = model.predict(test[predictors])
        predictions = pd.Series(predictions, index=test.index, name = "Predictions")
        combined = pd.concat([test["Target"], predictions], axis=1)
        return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(goog, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

Adding predictors to our model

```python
horizons = [2, 5, 50, 250, 1000]

new_predictors = []

for horizon in horizons:
    rolling_averages = goog.rolling(horizon).mean()
    
    ratio_column = f"Close Ratio_{horizon}"
    goog[ratio_column] = goog["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    goog[trend_column] = goog.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
goog = goog.dropna()
```

```python
goog
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model): 
        model.fit(train[predictors], train["Target"])
        predictions = model.predict_proba(test[predictors])[:,1]
        predictions[predictions >= .6] = 1
        predictions[predictions < .6] = 0
        predictions = pd.Series(predictions, index=test.index, name = "Predictions")
        combined = pd.concat([test["Target"], predictions], axis=1)
        return combined
```

```python
predictions = backtest(goog, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
